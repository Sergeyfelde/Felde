```
using System.Reflection;

Assembly asm = Assembly.LoadFrom("core.dll");
Console.WriteLine($"Полное имя сборки: {asm.FullName}");
Type[] types = asm.GetTypes();
Console.WriteLine($"Типы:");
Console.WriteLine($"\t╭―――――――――――――――――――");
Console.WriteLine($"\t│Классы:");
types.ToList().Where(ty => ty.IsClass == true).ToList().ForEach(t => 
{
    Console.WriteLine($"\t│Имя: {t.Name}");
    Console.WriteLine($"\t│\t╭―――――――――――――――――――");
    Console.WriteLine($"\t│\t│Публичные методы:");
    t.GetMethods().Where(me => me.IsPublic).ToList().ForEach(m => 
    {
        Console.WriteLine($"\t│\t│Имя:{m.Name}\t│Тип Возвращаемого значения: {m.ReturnType}");
        Console.WriteLine($"\t│\t│\t╭―――――――――――――――――――");
        Console.WriteLine($"\t│\t│\t│Параметры:");
        m.GetParameters().ToList().ForEach(p => 
        {
            Console.WriteLine($"\t│\t│\t|Имя:{p.Name}\t│Тип: {p.ParameterType}");
        }
        );
        Console.WriteLine($"\t│\t│\t╰―――――――――――――――――――");
    }
    );
    Console.WriteLine($"\t│\t╰―――――――――――――――――――");
}
);
Console.WriteLine($"\t╰―――――――――――――――――――");
Console.WriteLine($"\t╭―――――――――――――――――――");
Console.WriteLine($"\t│Интерфейсы:");
types.ToList().Where(ty => ty.IsInterface == true).ToList().ForEach(t => 
{
    Console.WriteLine($"\t│Имя: {t.Name}");
    Console.WriteLine($"\t│\t╭―――――――――――――――――――");
    Console.WriteLine($"\t│\t│Публичные методы:");
    t.GetMethods().Where(me => me.IsPublic).ToList().ForEach(m => 
    {
        Console.WriteLine($"\t│\t│Имя:{m.Name}\t│Тип Возвращаемого значения: {m.ReturnType}");
        Console.WriteLine($"\t│\t│\t╭―――――――――――――――――――");
        Console.WriteLine($"\t│\t│\t│Параметры:");
        m.GetParameters().ToList().ForEach(p => 
        {
            Console.WriteLine($"\t│\t│\t|Имя:{p.Name}\t│Тип: {p.ParameterType}");
        }
        );
        Console.WriteLine($"\t│\t│\t╰―――――――――――――――――――");
    }
    );
    Console.WriteLine($"\t│\t╰―――――――――――――――――――");
}
);
Console.WriteLine($"\t╰―――――――――――――――――――");
```
