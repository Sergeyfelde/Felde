```
using System.Reflection;

Assembly asm = Assembly.LoadFrom("core.dll");
Console.WriteLine(asm.FullName);
Type[] types = asm.GetTypes();
Console.WriteLine($"\tТипы:");
types.ToList().ForEach(t => 
{
    Console.Write($"\t Имя: {t.Name}");
    if (t.IsClass) Console.WriteLine($"\tКласс");
    if (t.IsInterface) Console.WriteLine($"\tИнтерфейс");
    Console.WriteLine($"\t\t╭―――――――――――――――――――");
    Console.WriteLine($"\t\t│Публичные методы:");
    t.GetMethods().ToList().ForEach(m => 
    {
        if (m.IsPublic) 
        {
            Console.WriteLine($"\t\t│Имя:{m.Name}\t│Тип Возвращаемого значения: {m.ReturnType}");
            Console.WriteLine($"\t\t│\t╭―――――――――――――――――――");
            Console.WriteLine($"\t\t│\t│Параметры:");
            m.GetParameters().ToList().ForEach(p => 
            {
                Console.WriteLine($"\t\t│\t|Имя:{p.Name}\t│Тип: {p.ParameterType}");
            }
            );
            Console.WriteLine($"\t\t│\t╰―――――――――――――――――――");
        };
    }
    );
    Console.WriteLine($"\t\t╰―――――――――――――――――――");
}
);
```
