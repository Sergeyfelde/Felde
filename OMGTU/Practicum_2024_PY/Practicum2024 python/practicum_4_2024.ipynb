```python
import numpy as np
import matplotlib.pyplot as plt

#1
print("П.1")


def y(x):
    return np.sin(np.sqrt(2*x))


a, b = 0, 1  # границы интегрирования
num_points = 100

os_x = np.linspace(a, b, num_points)
os_y = y(os_x)
print("Значения x:", *os_x)
print("Значения y:", *os_y)
```

```python
#2
print("П.2")

plt.figure(figsize=(10,6))
plt.plot(os_x, os_y, label="y(x) = sin(√2x),", color="b")

plt.title("График функции $y(x) = sin(√2x)$", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.legend(fontsize=12)

plt.grid(True)
plt.show()
```

```python
#3
print("П.3")

plt.figure(figsize=(10,6))
plt.scatter(os_x, os_y, label="y(x) = sin(√2x)", color=(50/255, 122/255, 2/255))

plt.title("График функции $y(x) = sin(√2x)$", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.legend(fontsize=12)

plt.grid(True)
plt.show()
```

```python
#4
print("П.4")
distrib_uniform = np.random.randint(1, 101, 100)
distrib_normal = np.round(np.random.normal(50, 10, 100)).astype(int)
# Настройка графиков
plt.figure(figsize=(12, 6))

# Первая гистограмма для равномерного распределения
plt.subplot(1, 2, 1)  # 1 строка, 2 колонки, 1-й график
plt.hist(distrib_uniform, bins=10, color='blue') #bins - значение кол-ва разбиений
plt.title('Гистограмма равномерного распределения')
plt.xlabel('Значения')
plt.ylabel('Частота')

# Вторая гистограмма для нормального распределения
plt.subplot(1, 2, 2)  # 1 строка, 2 колонки, 2-й график
plt.hist(distrib_normal, bins=6, color='green') #bins - значение кол-ва разбиений
plt.title('Гистограмма нормального распределения')
plt.xlabel('Значения')
plt.ylabel('Частота')

# Показать графики
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


#5
print("П.5")
# Генерация выборки из равномерного распределения (50 целых чисел от 1 до 5)
distrib_uniform_5 = np.random.randint(1, 5, 50)

# Подсчет количества вхождений каждого числа
values, counts = np.unique(distrib_uniform_5, return_counts=True)

# Цвета для диаграмм
colors = [(255/255, 153/255, 153/255), (102/255, 179/255, 255/255), (153/255, 255/255, 153/255), (255/255, 204/255, 153/255)]  # Цвета для круговой и столбчатой диаграмм

# Построение круговой диаграммы
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # 1 строка, 2 колонки, 1-й график
plt.pie(counts, labels=values, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Круговая диаграмма распределения чисел')

# Построение столбчатой диаграммы
plt.subplot(1, 2, 2)  # 1 строка, 2 колонки, 2-й график
plt.bar(values, counts, color=colors)
plt.xlabel('Числа')
plt.ylabel('Количество вхождений')
plt.title('Столбчатая диаграмма распределения чисел')
plt.xticks(values)  # Подписи для столбцов

# Показать графики
plt.tight_layout()
plt.show()
```

```python

#6
print("П.6")

def objective(x):
    return (x[0] + 3)**3 + (x[1] - 3)**2
#def objective(x):
#    return (x[0] - 3)**2 + (x[1] - 1)**2 #2 вариант
# Создание сетки для 3D графика
x1 = np.linspace(-100, 100, 100)
x2 = np.linspace(-100, 100, 100)
X1, X2 = np.meshgrid(x1, x2) # Создание координатных сеток

# Вычисление значений целевой функции для каждого (x1, x2)
Z = objective([X1, X2])

# Построение 3D графика
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')  # Установите цветовую карту

# Установка меток осей
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Значение целевой функции', labelpad=1)

plt.title('3D график целевой функции')
ax.view_init(elev=12, azim=120)
plt.show()
```
