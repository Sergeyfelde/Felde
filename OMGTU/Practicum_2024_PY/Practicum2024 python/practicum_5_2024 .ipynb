```python
import pandas as pd

#1
print("П.1")

data = {
    'Name': ['Strom, Mrs. Wilhelm (Elna Matilda Persson)', 'Navratil, Mr. Michel ("Louis M Hoffman")', 'Minahan, Miss. Daisy E'],
    'Age': [29, 36.5, 33],
    'Sex': ['female', 'male', 'female']
}

df_1 = pd.DataFrame(data)
df_1
```

```python
#2
print("П.2")

titanic_df_1 = pd.read_csv('titanic_csv.csv', sep=';')
titanic_df_1
```

```python
#3
print("П.3")

titanic_df_2 = pd.read_csv('https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt', sep=',')
titanic_df_2
```

```python
#4
print("П.4")
import pandas as pd

titanic_df_1.columns = titanic_df_1.columns.str.upper()
titanic_df_2.columns = titanic_df_1.columns.str.upper()

combined_df = pd.concat([titanic_df_1, titanic_df_2])

unique_df = combined_df.drop_duplicates()

unique_df
```

```python

#5
print("П.5")

unique_df.set_index('PASSENGERID', inplace=True)

df_sorted = unique_df.sort_index()

df_sorted
```

```python

#6
print("П.6")

df_sorted.info()

df_sorted.describe()
```

```python
#7
print("П.7")


row_0 = df_sorted.iloc[0]
row_2 = df_sorted.loc[2]  


df_sorted.iloc[0] = row_2


df_sorted.loc[2] = row_0

df_sorted
```

```python
#8
print("П.8")

df_sorted['SEX'] = df_sorted['SEX'].map({'female': 'f', 'male': 'm'})

df_sorted
```

```python
#9
print("П.9")

ticket_counts = df_sorted.groupby('TICKET').count()

ticket_counts_6_or_more = ticket_counts[ticket_counts['NAME'] >= 6]

df_sorted[df_sorted['TICKET'].isin(ticket_counts_6_or_more.index)]
```

```python
#10
print("П.10")

cabin_list_p1 = df_sorted[(df_sorted['NAME']).isin(df_1['Name'])]
people_list_w_p1 = df_sorted[(df_sorted['CABIN']).isin(cabin_list_p1['CABIN'])]
people_list_w_p1
```

```python
#11
print("П.11")

import math

df_sorted['BIRTHYEAR'] = df_sorted['AGE'].apply(lambda age: math.floor(1912.333 - age) if age >= 0 else None).astype('Int64')
df_sorted
```

```python
#12
print("П.12")

cabin_counts = df_sorted.groupby('CABIN')['NAME'].count()
df_sorted['COMPANION'] = df_sorted['CABIN'].apply(lambda cabin: cabin_counts[cabin] if cabin in cabin_counts.index else 0) 
df_sorted
```

```python
#13
print("П.13")

row_0 = df_sorted.iloc[0]  
row_1 = df_sorted.loc[1]   

df_sorted.iloc[0] = row_1


df_sorted.loc[1] = row_0

df_sorted
```

```python
#14
print("П.14")

df_sorted.to_csv('titanik_2.csv')
```
